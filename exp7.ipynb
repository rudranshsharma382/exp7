```python
#Factorial of a number
n = int(input("Enter a number: "))
fact = 1
while n > 0:
  fact *= n
  n -= 1

print("factorial=",fact)
```

```python
#Fibonacci Series using while loop

n = int(input("Enter the number of terms n: "))
a = 0
b = 1
i = 1
while i <= n:
  print(a,end=" ")
  c = a+b
  a = b
  b = c
  i+=1
```

```python
#Reverse a number
num = int(input("Enter a number: "))
rev = 0
while num > 0:
          digit = num % 10
          rev = rev * 10 + digit
          num //= 10

          print("Reverse:",rev)
```

```python
# check palidrome number/string
num = int(input("enter a number:"))
temp = num
rev = 0  # Initialize rev
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

# Check if palindrome after the loop finishes
if temp == rev :
    print("palindrome")
else:
    print("Not Palindrome")
```

```python
s = "madam"
i,j = 0,len(s) - 1
is_palindrome = True
while i < j:
    if s[i] != s[j]:
        is_palindrome = False
        break
    i += 1
    j -= 1

if is_palindrome:
    print("yes")
else:
    print("no")
```

```python
# check pelindrome number
st = input("Enter a string: ")
rev = st[::-1]
if st==rev:
  print("Palindrome")
else:

  print("Not Palindrome")
```

```python
#Count digit in number
num = int(input("Enter a number: "))
count = 0
while num > 0:
    count += 1
    num //= 10
    print("Number of digits =", count)
```

```python
#Search an element in a list
nums = [10,20,30,40,50]
key = int(input("Enter the element to search: "))
i = 0
while i < len(nums):
  if nums[i]== key :
    print("element found at index", i)
    break
  i += 1
else:
  print("element not found")
```
